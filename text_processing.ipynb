```python
import nltk
import spacy
import pandas as pd
import numpy as np
import string
import re
import timeit
import contractions
nltk.download('stopwords')
nltk.download('punkt_tab')
spacy.cli.download("en_core_web_sm")
```

## Tokenization


```python
def tokenization(input_text):
    """
    Function to test tokenization with NLTK and spaCy.
    - output - tuple
    """
    nlp = spacy.load('en_core_web_sm')
    tokens_nltk = nltk.word_tokenize(input_text)
    doc_spacy = nlp(input_text)
    tokens_spacy = [token.text for token in doc_spacy]
    return tokens_nltk, tokens_spacy
```

```python
tokens_nltk, tokens_spacy = tokenization(
"Alright, I've been thinking. When life gives you lemons," 
"don't make lemonade - make life take the lemons back! Get mad!" 
"I don't want your damn lemons, what am I supposed to do with these?" 
"Demand to see life's manager. Make life rue the day it thought it could give Cave Johnson lemons." 
"Do you know who I am? I'm the man who's gonna burn your house down!" 
"With the lemons. I'm going to to get my engineers to invent a combustible lemon that burns your house down!" 
)
print(f"NLTK Tokens: {tokens_nltk} spaCy tokens: {tokens_spacy}")
print(list(set(tokens_nltk) ^ set(tokens_spacy)))
```

## Stop Words Removal

```python
def stop_words_removal(tokens: str):
    """
    Function to test stop words removal with nltk and spacy.
    - output - tuple  
    - tokens | str
    """
    nlp = spacy.load('en_core_web_sm')
    stop_words = set(nltk.corpus.stopwords.words('english'))
    tokens_spacy = ' '.join(tokens)
    doc = nlp(tokens_spacy)
    filtered_tokens_spacy = [token.text for token in doc if not token.is_stop]
    filtered_tokens_nltk = [w for w in tokens if not w in stop_words]
    return filtered_tokens_spacy, filtered_tokens_nltk
```

```python
filtered_tokens_spacy, filtered_tokens_nltk = stop_words_removal(tokens_spacy)
print(stop_words_removal(tokens_spacy))
print(stop_words_removal(tokens_nltk))
```

## Normalization

```python
def normalization(filtered_tokens: any):
    '''
    Function to normalize the outputs from tokenization and stop words removal.
    - output - string
    '''

    normalized_tokens = [re.sub(r'[^a-zA-Z0-9]', '', token.lower()) for token in filtered_tokens]
    normalized_tokens = [token for token in normalized_tokens if token]
    return normalized_tokens

```

```python
print(normalization(filtered_tokens_spacy))
normalized_tokens_spacy = normalization(filtered_tokens_spacy)
```

## Lemmatization

```python
def lemmatization(filtered_tokens: any):
    '''
    Function to lemmatize the outputs from tokenization and stop words removal.
    - output - string
    '''
    nlp = spacy.load('en_core_web_sm')
    tokens_spacy = ' '.join(filtered_tokens)
    doc = nlp(tokens_spacy)
    lemmatized_tokens_spacy = [token.lemma_ for token in doc]
    return lemmatized_tokens_spacy
```

```python
print(lemmatization(normalized_tokens_spacy))
lemmatized_tokens_spacy = lemmatization(normalized_tokens_spacy)
```

## Stemming

```python
def stemming(filtered_tokens: any):
    '''
    Function to stem the outputs from tokenization and stop words removal.
    - output - string
    '''
    porter_stemmer = nltk.PorterStemmer()
    stemmed_tokens_nltk = [porter_stemmer.stem(token) for token in filtered_tokens]
    return stemmed_tokens_nltk
```

```python
print(stemming(normalized_tokens_spacy))
stemmed_tokens_nltk = stemming(normalized_tokens_spacy)
```

## Time and Output Comparision

```python
lemma_time = timeit.timeit(lambda: lemmatization(normalized_tokens_spacy), number=100)
stem_time = timeit.timeit(lambda: stemming(normalized_tokens_spacy), number=100)

print(f"Lemmatization Execution Time: {lemma_time:.4f} seconds")
print(f"Stemming Execution Time:      {stem_time:.4f} seconds")
print(f"Speed Difference:             Stemming is roughly {lemma_time / stem_time:.1f}x faster")
```

## Proper Pre-processing order w/ Normalization first

```python
# Need to handle contractions in the normalization step so we get good results from lemmatization.
#Normalization
#Tokenization
#Stop Words Removal
#Lemmatization

def updated_normalization(input_text):
    '''
    Function to normalize the text before tokenization and stop words removal.
    Accepts a string or a list of strings (lines).
    - output - list of normalized tokens
    '''
    if isinstance(input_text, list):
        input_text = ' '.join(input_text)
    # Expand contractions first
    input_text = contractions.fix(input_text)
    # Tokenize and normalize
    normalized_tokens = [re.sub(r'[^a-zA-Z0-9]', '', token.lower()) for token in input_text.split()]
    normalized_tokens = [token for token in normalized_tokens if token]
    return normalized_tokens
```

```python
text_lines = [
    "Alright, I've been thinking. When life gives you lemons,",
    "don't make lemonade - make life take the lemons back! Get mad!",
    "I don't want your damn lemons, what am I supposed to do with these?",
    "Demand to see life's manager. Make life rue the day it thought it could give Cave Johnson lemons.",
    "Do you know who I am? I'm the man who's gonna burn your house down!",
    "With the lemons. I'm going to get my engineers to invent a combustible lemon that burns your house down!"
]

cleaned_text = updated_normalization(text_lines)
```

```python
tokens_nltk, tokens_spacy = tokenization(' '.join(cleaned_text))
print(f"NLTK Tokens: {tokens_nltk} spaCy tokens: {tokens_spacy}")
```

```python
filtered_tokens_spacy, filtered_tokens_nltk = stop_words_removal(tokens_spacy)
print(stop_words_removal(tokens_spacy))
print(stop_words_removal(tokens_nltk))
```

```python
print(lemmatization(filtered_tokens_spacy))
lemmatized_tokens_spacy = lemmatization(filtered_tokens_spacy)
```
