```python
import nltk
import spacy
import pandas as pd
import numpy as np
import string
import re
nltk.download('stopwords')
spacy.cli.download("en_core_web_sm")
```

## Tokenization


```python
def tokenization(input_text):
    """
    Function to test tokenization with NLTK and spaCy.
    - output - tuple
    """
    nlp = spacy.load('en_core_web_sm')
    tokens_nltk = nltk.word_tokenize(input_text)
    doc_spacy = nlp(input_text)
    tokens_spacy = [token.text for token in doc_spacy]
    return tokens_nltk, tokens_spacy
```

```python
tokens_nltk, tokens_spacy = tokenization("I am giving this review 3 out of 5 stars. Not great but not bad either.")
print(f"NLTK Tokens: {tokens_nltk} spaCy tokens: {tokens_spacy}")
```

## Stop Words Removal

```python
def stop_words_removal(tokens: str):
    """
    Function to test stop words removal with nltk and spacy.
    - output - tuple  
    - tokens | str
    """
    nlp = spacy.load('en_core_web_sm')
    stop_words = set(nltk.corpus.stopwords.words('english'))
    tokens_spacy = ' '.join(tokens)
    doc = nlp(tokens_spacy)
    filtered_tokens_spacy = [token.text for token in doc if not token.is_stop]
    filtered_tokens_nltk = [w for w in tokens if not w in stop_words]
    return filtered_tokens_spacy, filtered_tokens_nltk
```

```python
filtered_tokens_spacy, filtered_tokens_nltk = stop_words_removal(tokens_spacy)
print(stop_words_removal(tokens_spacy))
print(stop_words_removal(tokens_nltk))
```

## Normalization

```python
def normalization(filtered_tokens: any):
    '''
    Function to normalize the outputs from tokenization and stop words removal.
    - output - string
    '''

    normalized_tokens = [re.sub(r'[^a-zA-Z0-9]', '', token.lower()) for token in filtered_tokens]
    normalized_tokens = [token for token in normalized_tokens if token]
    return normalized_tokens

```
